# Nomenclator

```python
with open ('data/nomenclator.json') as f:
    nomenclator = json.load(f)
nomenclator_df = pd.DataFrame(nomenclator["equipaments"])
geometry = gpd.GeoSeries.from_wkt(nomenclator_df["WKT"], on_invalid="ignore")

print("Invalid geometries:", geometry.isna().sum())

geodf = gpd.GeoDataFrame(
    nomenclator_df.copy(),
    geometry=geometry,
    crs="EPSG:4326",
)

geodf = geodf.dropna(subset=["geometry"])
```

```python
shape_layer = alt.Chart(geodf).mark_geoshape(size = 0,filled = False,color = 'black')
bus1_layer = alt.Chart(geo_df_points_1).mark_geoshape(size=0, color='red').encode(tooltip=alt.Tooltip('label:N', title='Bus Stop'))
bus2_layer = alt.Chart(geo_df_points_2).mark_geoshape(size=0, color='blue').encode(tooltip=alt.Tooltip('label:N', title='Bus Stop'))
(shape_layer+ bus1_layer+ bus2_layer).properties(
    width=600,
    height=600
)
```
